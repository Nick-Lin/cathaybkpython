```python
import time
for row in page.head().iterrows():
    print(row[1]['SessionNumber'], row[1]['EventTimestamp'])
    print(time.mktime(row[1].EventTimestamp.timetuple()))
```

```python
session_dic = {}
for row in page2.iterrows():
    session = row[1]['SessionNumber']
    eventtime =  time.mktime(row[1].EventTimestamp.timetuple())
    pageLocation =  row[1]['PageLocation']
    if session not in session_dic:
        session_dic[session] = [pageLocation]
    else:
        session_dic[session].append(pageLocation)
```

```python
transactions = tuple([tuple(session_dic[session]) for session in session_dic])
```

```python
len(transactions)
```

```python
from pymining import itemmining, assocrules, perftesting
#transactions = perftesting.get_default_transactions()
relim_input = itemmining.get_relim_input(transactions)
item_sets = itemmining.relim(relim_input, min_support=100)
rules = assocrules.mine_assoc_rules(item_sets, min_support=100, min_confidence=0.3)
#rules
```

```python
for r in rules:
    if len([l for l in r[0] if 'card_apply' in l]) > 0:
        print(r)
```
