## 讀取iris

```python
from sklearn.datasets import load_iris
iris = load_iris()
```

## 探索iris 資料集 

```python
iris.data.shape
```

```python
iris.target
```

## 繪製資料 (根據Petal.Length, Petal.Width)

```python
%pylab inline
import matplotlib.pyplot as plt
plt.xlabel = 'Petal.Length'
plt.ylabel = 'Petal.Width'
plt.figure(figsize=(12, 12))
plt.scatter(iris.data[:,2], iris.data[:,3])
plt.show()
```

## 實際結果

```python
%pylab inline
import matplotlib.pyplot as plt
plt.xlabel = 'Petal.Length'
plt.ylabel = 'Petal.Width'
plt.figure(figsize=(12, 12))
plt.scatter(iris.data[:,2], iris.data[:,3],c=iris.target)
plt.show()
```

## 使用sklearn 進行預測

```python
from sklearn.cluster import KMeans
k_means = KMeans(n_clusters = 3)
pred = k_means.fit_predict(iris.data)
```

```python
pred
```

## 根據預測結果繪製資料

```python
%pylab inline
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 12))
plt.scatter(iris.data[:,2], iris.data[:,3],c=pred)
plt.show()
```

## 比較原先的資料與預測資料

```python
plt.figure(figsize=(12, 12))
plt.subplot(221)
plt.scatter(iris.data[:,2], iris.data[:,3],c=pred)
plt.subplot(222)
plt.scatter(iris.data[:,2], iris.data[:,3],c=iris.target)
```

## 自由調整群數

```python
from sklearn.cluster import KMeans
k_means2 = KMeans(n_clusters = 5)
pred = k_means2.fit_predict(iris.data)
plt.figure(figsize=(12, 12))
plt.scatter(iris.data[:,2], iris.data[:,3],c=pred)
plt.show()
```
